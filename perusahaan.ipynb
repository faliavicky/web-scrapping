```python
driver = Chrome()
driver.get('https://klikalamat.com/edukasi-kursus/')
```

```python
import time

time.sleep(10)
```

```python
page = 2
nama = []
nomor = []
alamat = [] 
```

```python
while True:
        time.sleep(10)
        html = driver.page_source
        soup = BeautifulSoup(html, 'html.parser')
        web = soup.find_all("section", {'class':'search_result_listing list'})
        web_nama = soup.find_all(attrs={'class':'entry-title'})
        web_nomor = soup.find_all('p',attrs={'class':'phone'})
        web_alamat = soup.find_all(attrs={'class':'address'})

        for element_nama in web_nama:
                nama.append(element_nama.text)

        for element_nomor in web_nomor:
                nomor.append(element_nomor.text)

        for element_alamat in web_alamat:
                alamat.append(element_alamat.text)

        if len(driver.find_elements(By.CLASS_NAME, 'page-numbers')) > 0:
                url = "https://klikalamat.com/edukasi-kursus/page/{}/".format(page)
                driver.get(url)
                page += 1
        else:
              break
        
'''
        for results in soup:
            nama.append(soup.find(attrs={'class':'entry-title'}).get_text().strip())
            nomor.append(soup.find('p', attrs={'class':'phone'}).get_text().strip())
            alamat.append(soup.find(attrs={'class':'address'}).get_text().strip())'''

        
```
